## What is Pandas?
- **Pandas** is a powerful Python library for data analysis and manipulation
- It provides two main data structures: **Series** (1D) and **DataFrame** (2D)
- Think of a DataFrame like an Excel spreadsheet in Python

## **Lesson 1: Loading Data**

The first step in any data analysis is loading your data into pandas.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv(f"{path}/online_sales_dataset.csv")

# Display basic info
print(f"Dataset shape: {df.shape}")
print(f"\nFirst few rows:")
print(df.head())
```

## 2️⃣ **Lesson 2: Basic DataFrame Information**

Let's explore what's in our dataset!

### 📝 What we learned:
- `pd.read_csv()` - reads CSV files into a DataFrame
- `.shape` - shows (rows, columns)
- `.head()` - displays first 5 rows (default)

```python
# Get column names
print("Column Names:")
print(df.columns.tolist())
print("\n" + "="*50 + "\n")

# Get data types of each column
print("Data Types:")
print(df.dtypes)
print("\n" + "="*50 + "\n")

# Get summary information
print("Dataset Info:")
df.info()
```

## 3️⃣ **Lesson 3: Viewing Data**

Different ways to look at your data.

### 📝 What we learned:
- `.columns` - lists all column names
- `.dtypes` - shows data type of each column (int64, float64, object, etc.)
- `.info()` - gives overview: number of entries, non-null values, memory usage

```python
# View first 5 rows
print("First 5 rows:")
print(df.head())
print("\n" + "="*50 + "\n")

# View last 5 rows
print("Last 5 rows:")
print(df.tail())
print("\n" + "="*50 + "\n")

# View random 5 rows
print("Random 5 rows:")
print(df.sample(5))
print("\n" + "="*50 + "\n")

# View specific number of rows
print("First 3 rows:")
print(df.head(3))
```

### 📝 What we learned:
- `df['column']` - select single column → returns **Series**
- `df[['col1', 'col2']]` - select multiple columns → returns **DataFrame**
- Note: Double brackets [[ ]] for multiple columns!

```python
# First, let's see what columns we have
print("Available columns:", df.columns.tolist()[:5])  # Show first 5 columns
print("\n" + "="*50 + "\n")

# Select a single column (returns a Series)
print("Single Column Selection:")
first_col = df.columns[0]  # Get the first column name
print(df[first_col].head())
print(f"Type: {type(df[first_col])}")
print("\n" + "="*50 + "\n")

# Select multiple columns (returns a DataFrame)
print("Multiple Column Selection:")
cols = df.columns[:2].tolist()  # Get first 2 columns
print(df[cols].head())
print(f"Type: {type(df[cols])}")
```

## 5️⃣ **Lesson 5: Selecting Columns**

How to access specific columns from your DataFrame.

### 📝 What we learned:
- `.describe()` - shows count, mean, std, min, max, quartiles for numeric columns
- `.count()` - counts non-null values in each column

```python
# Statistical summary of numerical columns
print("Statistical Summary:")
print(df.describe())
print("\n" + "="*50 + "\n")

# Count of values in each column
print("Value Counts:")
print(df.count())
```
